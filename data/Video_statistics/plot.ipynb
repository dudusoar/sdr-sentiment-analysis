```python
import pandas as pd

# Load the uploaded CSV file
file_path = 'Publish_Count_by_Year.csv'
data = pd.read_csv(file_path)

# Display the data to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt

# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(data['year'], data['count'])

# Add text labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom')

# Modify the x-axis label for the last bar
plt.xticks(ticks=data['year'], labels=[str(year) if year != data['year'].iloc[-1] else 'Apr. 2023' for year in data['year']], fontsize=12)
plt.yticks(fontsize=12)

# Remove the title
plt.title('')

# Display the plot
plt.tight_layout()
plt.show()
plt.savefig('Publish count by year.png')
```
